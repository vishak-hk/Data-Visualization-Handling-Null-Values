### 1. Import the necessary libraries: Start by importing the required libraries, including Matplotlib, NumPy, and pandas.

```python
# import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Load the dataset: Load the Iris dataset(load_iris) from sklearn.datasets convert it into a pandas DataFrame.
### The dataset contains 150 rows and 5 columns (4 features and 1 target variable).

```python
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a Pandas DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add a 'target' column for the target labels
iris_df['species'] = iris.target
```

### 3. Line Plot: Create a line plot to visualize the trend in the data over time. Since this is not a time-series dataset, create a line plot for each feature against the target variable(species).
### Customize the plot by adding a title, labels for the x and y-axis, and a legend.

```python
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

for feature in features:
    plt.plot(iris_df['species'], iris_df[feature], marker='o', linestyle='-')

# Display all the plots together
plt.xlabel('Species (0: Setosa, 1: Versicolor, 2: Virginica)')
plt.ylabel('Features')
plt.title('Features vs Species')
plt.legend(features)
plt.show()
```

### 6. Bar Chart: Create a bar chart to visualize the frequency or count of a categorical variable.
### Create a bar chart for the target variable (species), showing the count of each species. Customize the plot by adding a title, labels for the x and y-axis, and changing the color scheme.

```python
species_count = iris_df['species'].value_counts()
print(species_count)

custom_palette = ["green", "yellow", "red"]

sns.set_style("whitegrid")
sns.barplot(x=species_count.index, y=species_count.values, palette=custom_palette)

plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Species Count Bar Chart')

plt.show()
```

### 7. Heatmap: Create a heatmap to visualize the correlation between multiple variables.
### Create a heatmap showing the correlation matrix between the features. Customize the plot by adding a title and changing the color scheme.

```python
correlation_matrix = iris_df.corr()

custom_palette = sns.color_palette("coolwarm")

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap=custom_palette)

plt.title('Correlation Heatmap')

plt.show()
```

### 8. Bonus: Choose a different dataset and create a unique visualization using Matplotlib. Be creative!

```python
wind_directions = ['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW']
frequency = [25, 10, 35, 18, 30, 12, 28, 15]

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)

theta = np.linspace(0, 2 * np.pi, len(wind_directions), endpoint=False)

ax.bar(theta, frequency, width=0.4, alpha=0.6, color='blue')

ax.set_xticks(theta)
ax.set_xticklabels(wind_directions)
ax.set_title('Wind Direction Distribution', fontsize=14, pad=20)

legend_labels = [f'{dir} ({freq}%)' for dir, freq in zip(wind_directions, frequency)]
ax.legend(legend_labels, loc='upper right')

plt.show()
```

### 1. Remove rows with missing values in the 'Embarked' column, as there were only two missing values and they can be safely dropped.


```python
titanic_df.dropna(subset='Embarked', inplace=True)
```

```python
titanic_df.info()
```

### 2. Fill missing values in the 'Age' column with the median age.

```python
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
```

```python
titanic_df.info()
```

```python
titanic_df.isnull().sum()
```

### 3. Create a new category "Unknown" for the 'Cabin' column to represent missing values

```python
titanic_df['Cabin'].fillna('Unknown', inplace=True)
```

```python
titanic_df.info()
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df.columns
```

```python
titanic_df['Cabin'].isnull().sum()
```

```python
survived_values = []

for index, value in titanic_df['Cabin'].items():
    if value == 'Unknown':
        survived_values.append(titanic_df.loc[index, 'Survived'])

count_survived = survived_values.count(1)
count_not_survived = survived_values.count(0)

print("Count of 'Survived' = 1:", count_survived)
print("Count of 'Survived' = 0:", count_not_survived)
```
